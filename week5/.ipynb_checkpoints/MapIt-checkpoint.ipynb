```python
#importing the library as shortname 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
classes_inventory = {
    'major related required classes': 29, 
    'major related electives': 6, 
    'minor required classes': 2, 
    'general electives': 7, 
    }
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.bar(range(len(classes_inventory)), list(classes_inventory.values()), align='center', color=['#DFDBDA', '#CDBC7A', '#2D65A0', '#76BEB1'])
plt.xticks(range(len(classes_inventory)), list(classes_inventory.keys()), rotation = 45)

#plt.legend()
plt.xlabel('Course Categories')
plt.ylabel('# of Classes')
plt.title('My Academic Journey')
plt.show()
```

```python
# create data
x = ['major related required classes','major related electives','minor required classes','general electives']
y1 = np.array([16, 9, 2, 0])
y2 = np.array([10, 2, 1, 0])
y3 = np.array([7, 3, 3, 7])
 
# plot bars in stack manner
plt.bar(x, y1, color='#76BEB1')
plt.bar(x, y2, bottom=y1, color='#CDBC7A')
plt.bar(x, y3, bottom=y1+y2, color='#DFDBDA')
plt.xticks(rotation = 25)

#plt.legend()
plt.xlabel('Course Categories')
plt.ylabel('# of Classes')
plt.title('My Academic Journey')
plt.legend(['satisfied','somewhat','not so much'])

plt.show()
```

```python
# pie plot 
# slices dictate how big your slices are

# Creating dataset
relevance = ['Relevant','Somewhat','Not Much']
data = [23, 19, 14]
colors = ( '#F05B38', '#2D65A0', '#6199AD')
 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = relevance, colors = colors)


 
# show plot
plt.show()
```
